##  End-to-end workflow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#features matrix
X = heart_disease.drop("target",axis=1)

#labels
Y = heart_disease["target"]
X
```

```python
#choosing ML model and hyperparameters
from sklearn.ensemble import RandomForestClassifier 
clf = RandomForestClassifier()

#default hyperparameters
clf.get_params()
```

```python
#fit model to training data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
clf.fit(X_train,Y_train)
```

```python
#make predictions with test data
Y_preds = clf.predict(X_test)
Y_preds
```

```python
#evaluate the model on training data and test data
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(Y_test,Y_preds))
```

```python
print(confusion_matrix(Y_test,Y_preds))
```

```python
accuracy_score(Y_test,Y_preds)
```

```python
#improve the model (e.g different n_estimators/number of trees)
for i in range(10,100,10):
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,Y_train)
    print(f'Training size: {i}%')
    print(f'Accuracy: {clf.score(X_test,Y_test)}')
    print()
```

```python
#save the model + load it
import pickle

pickle.dump(clf, open("random-forest-model.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random-forest-model.pkl", "rb"))
loaded_model.score(X_test,Y_test)
```

```python
#fill missing values with sklearn

car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing.head()
```

```python
car_sales_missing.dropna(subset=["Price"],inplace=True)
car_sales_missing.isna().sum()
```

```python
X = car_sales_missing.drop("Price",axis=1)
Y = car_sales_missing["Price"]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

cat_imputer = SimpleImputer(strategy="constant",fill_value="missing")
door_imputer = SimpleImputer(strategy="constant",fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

cat_features = ['Make','Colour']
num_features = ["Odometer (KM)"]
door_features = ["Doors"]

imputer = ColumnTransformer([
    ("cat_imputer",cat_imputer,cat_features),
    ('num_imputer',num_imputer,num_features),
    ('door_imputer',door_imputer,door_features)
])

filled_X = imputer.fit_transform(X)
filled_car_sales = pd.DataFrame(filled_X,columns=["Make",'Colour','Doors','Odometer (KM)'])
filled_car_sales.head()
```

## Choosing the right estimator

```python
#regression
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df
```

```python
housing_df["MedianVal"] = housing["target"]
housing_df.head()
```

```python
from sklearn.linear_model import Ridge

X = housing_df.drop("MedianVal",axis=1)
Y = housing_df["MedianVal"]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)

model = Ridge()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

X = housing_df.drop("MedianVal",axis=1)
Y = housing_df["MedianVal"]

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
#predicting with trained model (predict_proba would return prediction probability of all groups in classification)
Y_preds = model.predict(X_test)
```

```python
np.array(Y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_preds,Y_test)
```
