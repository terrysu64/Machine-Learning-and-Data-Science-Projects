##  End-to-end workflow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
#features matrix
X = heart_disease.drop("target",axis=1)

#labels
Y = heart_disease["target"]
X
```

```python
#choosing ML model and hyperparameters
from sklearn.ensemble import RandomForestClassifier 
clf = RandomForestClassifier()

#default hyperparameters
clf.get_params()
```

```python
#fit model to training data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
clf.fit(X_train,Y_train)
```

```python
#make predictions with test data
Y_preds = clf.predict(X_test)
Y_preds
```

```python
#evaluate the model on training data and test data
clf.score(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(Y_test,Y_preds))
```

```python
print(confusion_matrix(Y_test,Y_preds))
```

```python
accuracy_score(Y_test,Y_preds)
```

```python
#improve the model (e.g different n_estimators/number of trees)
for i in range(10,100,10):
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,Y_train)
    print(f'Training size: {i}%')
    print(f'Accuracy: {clf.score(X_test,Y_test)}')
    print()
```

```python
#save the model + load it
import pickle

pickle.dump(clf, open("random-forest-model.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random-forest-model.pkl", "rb"))
loaded_model.score(X_test,Y_test)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

X = car_sales.drop('Price',axis=1)
categorical_features = ['Make','Colour','Doors']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder='passthrough')

transformed_X = transformer.fit_transform(X)
transformed_X
```
