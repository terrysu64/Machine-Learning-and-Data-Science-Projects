## Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
plt.plot()
```

```python
plt.plot([1,2,3,4])
```

```python
x = [1,2,3,4]; y = [11,22,34,63]
plt.plot(x,y)
```

```python
#method 1
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#method 2
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
#method 3
fig, ax = plt.subplots()
ax.plot(x,y)
type(fig), type(ax)
```

## Matplotlib example workflow

```python
# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="sample plot", xlabel='x-axis',ylabel='y-axis')

# 5. save figure
fig.savefig("sample-plot.png")
```

## Making figures with Numpy

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2)
```

```python
#scatter plot
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
#bar plot
pets = {j:i for i,j in enumerate(['cat','dog','fish','amongus'])}
fig,ax = plt.subplots()
ax.bar(pets.keys(),pets.values())
```
