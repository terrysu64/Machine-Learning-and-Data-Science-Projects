## Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
plt.plot()
```

```python
plt.plot([1,2,3,4])
```

```python
x = [1,2,3,4]; y = [11,22,34,63]
plt.plot(x,y)
```

```python
#method 1
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#method 2
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
#method 3
fig, ax = plt.subplots()
ax.plot(x,y)
```
